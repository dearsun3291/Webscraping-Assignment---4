## 2. Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
def match_fixture(url): # defining function
    #initalizing driver & loading url
    driver = webdriver.Chrome(r"C:\Users\clear\Desktop\webdriver\chromedriver_win32\chromedriver.exe")
    driver.get(url)
    # Getting internation match fixtures page
    driver.get(driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down']").get_attribute('href'))
    # Preparing soup & scraping data
    soup = BeautifulSoup(driver.page_source, 'lxml')
    match_title = [x.text for x in soup.find_all('span',class_ ='u-unskewed-text fixture__format')]
    series = [x.text for x in soup.find_all('span',class_ ='u-unskewed-text fixture__tournament-label u-truncated')]
    place = [x.text for x in soup.find_all('p',class_ ='fixture__additional-info')]
    date = [x.text for x in soup.find_all('span',class_ ='fixture__datetime tablet-only')]
    driver.close()
    # Preparing dataframe & cleaning data
    df= pd.DataFrame({})
    df['match_title']=match_title
    df['series']=series
    df['place1']=place
    df['date1']=date
    dfp = df.place1.str.split('\n',n = 4, expand = True)
    dfp.columns = ['d1','d2','place','df4']
    dfp.drop(['d1','d2','df4'],axis =1,inplace =True)
    dfd= df.date1.str.split('\n',n = 4, expand = True)
    dfd.columns = ['d1','date','time','d3','d4']
    dfd.drop(['d1','d3','d4'],axis =1,inplace =True)
    df = pd.concat([df,dfp,dfd],axis =1)
    df.drop(['place1','date1'], axis =1, inplace = True)
    return(df)


match_fixture('https://www.bcci.tv/') # Calling function
```

## 5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
def github(url):
    driver = webdriver.Chrome(r"C:\Users\clear\Desktop\webdriver\chromedriver_win32\chromedriver.exe")
    driver.get(url)
    driver.find_element_by_xpath("//li[4][@class='d-block d-lg-flex flex-lg-nowrap flex-lg-items-center border-bottom border-lg-bottom-0 mr-0 mr-lg-3 edge-item-fix position-relative flex-wrap flex-justify-between d-flex flex-items-center ']/details/summary").click()
    menu=driver.find_elements_by_xpath("//a[@class='py-2 lh-condensed-ultra d-block Link--secondary no-underline f5 Bump-link--hover']")
    links=[]
    for i in menu:
        links.append(i.get_attribute('href'))


    driver.get(links[10])
    soup = BeautifulSoup(driver.page_source,'lxml')
    details = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
    links_ele=[]
    for i in details:
        links_ele.append(i.get_attribute('href'))

    titles= []
    Repository_description=[]
    Contributors=[]
    Language=[]
    for i in links_ele:
        driver.get(i)
        title= driver.find_elements_by_xpath("//a[@class='url fn']")
        for i in title:
            titles.append(i.text)

        try:
            delay = 2
            WebDriverWait(driver,delay).until(expected_conditions.presence_of_element_located((By.XPATH,"//p[@class='f4 mt-3']")))
            desc= driver.find_elements_by_xpath("//p[@class='f4 mt-3']")
            for i in desc:
                Repository_description.append(i.text)

        except TimeoutException:
            Repository_description.append('-')

        try:
            delay = 2
            WebDriverWait(driver,delay).until(expected_conditions.presence_of_element_located((By.XPATH,"//div[@class='BorderGrid BorderGrid--spacious']/div[5]/div/h2/a/span")))
            contri = driver.find_elements_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[5]/div/h2/a/span")
            for i in contri:
                Contributors.append(i.text)

        except TimeoutException:
            Contributors.append('-')

        Lange=[]
        try:
            delay = 5
            WebDriverWait(driver,delay).until(expected_conditions.presence_of_element_located((By.XPATH,"//div[@class='BorderGrid BorderGrid--spacious']/div[6]/div/ul/li/a/span[1]")))
            lang = driver.find_elements_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[6]/div/ul/li/a/span[1]")
            for i in lang:
                Lange.append(i.text)
        
        except TimeoutException:
            Lange.append('-')
                
        Language.append(Lange)

    driver.close()
    df = pd.DataFrame({})
    df['titles']=titles
    df['Repository_description']=Repository_description
    df['Contributors']=Contributors
    df['Language']=Language
    return(df)

github('https://github.com/')
```

## 6. Scrape the details of top 100 songs on billiboard.com.
Url = https://www.billboard.com/
You have to find the following details:
A) Song name
B) Artist name
C) Last week rank
D) Peak rank
E) Weeks on board
Note: - From the home page you have to click on the charts option then hot 100-page link through code.
    

```python
def songs(url):
    driver=webdriver.Chrome(r"C:\Users\clear\Desktop\webdriver\chromedriver_win32\chromedriver.exe")
    driver.get(url)
    delay = 10
    WebDriverWait(driver,delay).until(expected_conditions.presence_of_element_located((By.XPATH,"//div[@class='sticky header__sticky']/div[2]/header/div/ul/li[1]/a")))
    driver.find_element_by_xpath("//div[@class='sticky header__sticky']/div[2]/header/div/ul/li[1]/a").click()
    driver.get(driver.find_element_by_xpath("//a[@class='charts-landing__link charts-landing__video--background']").get_attribute('href'))
    song_name=[]
    artist=[]
    last_week_rank=[]
    peak_rank=[]
    weeks_on_board=[]

    WebDriverWait(driver,delay).until(expected_conditions.presence_of_element_located((By.XPATH,"//span[@class='chart-element__information']/span[1]")))
    table = driver.find_elements_by_xpath("//span[@class='chart-element__information']/span[1]")
    for i in table:
        song_name.append(i.text)

    table = driver.find_elements_by_xpath("//span[@class='chart-element__information']/span[2]")
    for i in table:
        artist.append(i.text)

    table = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
    for i in table:
        last_week_rank.append(i.text)

    table = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
    for i in table:
        peak_rank.append(i.text)


    table = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
    for i in table:
        weeks_on_board.append(i.text)

    driver.close()
    df = pd.DataFrame({})
    df['song_name']=song_name
    df['artist']=artist
    df['last_week_rank']=last_week_rank
    df['peak_rank']=peak_rank
    df['weeks_on_board']=peak_rank
    return(df)

songs('https://www.billboard.com/')
```

## 7. Scrape the details of Data science recruiters from naukri.com.
Url = https://www.naukri.com/
You have to find the following details:
A) Name
B) Designation
C) Company
D) Skills they hire for
E) Location
Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and 
click on search. All this should be done through code

```python
def naukri(url):
    driver=webdriver.Chrome(r"C:\Users\clear\Desktop\webdriver\chromedriver_win32\chromedriver.exe")
    driver.get(url)
    delay = 10
    WebDriverWait(driver,delay).until(expected_conditions.presence_of_element_located((By.XPATH,"//a[@title='Search Recruiters']")))
    driver.get(driver.find_element_by_xpath("//a[@title='Search Recruiters']").get_attribute('href'))
    driver.find_element_by_xpath("//input[@class='sugInp']").send_keys('Data Scientist')
    driver.find_element_by_id("qsbFormBtn").click()
    det = [x.text for x in driver.find_elements_by_xpath("//p[@class='highlightable']")]
    skills = [x.text for x in driver.find_elements_by_xpath("//div[@class='hireSec highlightable']")]
    driver.close()
    df = pd.DataFrame({})
    df['det']=det
    df['skills']=skills
    dft= df.det.str.split('\n',expand=True)
    dft.columns = ['Name','Designation','Company','Location']
    df = pd.concat([dft,df], axis =1)
    df.drop('det',axis=1,inplace=True)
    return(df)

naukri('https://www.naukri.com/jobs-in-india')
```

## 10. Details of Datasets from UCI machine learning repositories.
 Url = https://archive.ics.uci.edu/

A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute
G) Year
Note: - from the home page you have to go to the Show All Dataset page through code

```python
def uci(url):
    driver=webdriver.Chrome(r"C:\Users\clear\Desktop\webdriver\chromedriver_win32\chromedriver.exe")
    driver.get(url)
    time.sleep(3)
    link=[x.get_attribute('href') for x in driver.find_elements_by_xpath("//td/span[@class='whitetext']/a")]
    driver.get(link[4])
    time.sleep(3)
    name = [x.text for x in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td")]
    df = pd.DataFrame({})
    df["Dataset_name"]=name[0::7]
    df['Data_type']=name[1::7]
    df['Task']=name[2::7]
    df['Attribute_type']=name[3::7]
    df['No_of_instances']=name[4::7]
    df['No of attribute']=name[5::7]
    df['Year']=name[6::7]
    df.drop(labels=0,axis =0, inplace =True)
    driver.close()
    return(df)

uci('https://archive.ics.uci.edu/')
```
